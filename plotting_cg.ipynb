```python
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
import numpy as np
```

```python
df = pd.read_csv('results/col_Gen.csv')
#df = df[df['runtime'] < 2000]
df['normalized_trading'] = df['trading_fee'] / df['initial_budget']
df['gain_without_trading'] = (df['Final Value'] + df['total_trading_costs'] - df['initial_budget']) / df['Initial Value']
df['gain_lost_to_trading'] = (df['Gain'] - df['gain_without_trading']) * 100
df['gain_without_trading'] = df['gain_without_trading'] * 100
df['log_pct_variance'] = np.log(df['pct_variance'] + 1)
df['log_gain'] = np.log(df['Gain'] + 1)
df['% Change'] = df['Gain'] * 100
df['log_normalized_trading'] = np.log(df['normalized_trading'])
```
